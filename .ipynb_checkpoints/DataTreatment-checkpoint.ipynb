### Solve null elements problem

```python
movies_v2.notnull() #View null elements
```

```python
movies_v2.notnull().apply(pd.Series.value_counts) #Count null elements
```

```python
(movies_v2 != 0).apply(pd.Series.value_counts) #Count zero elements
```

```python
available = ((movies_v2 != 0) & (movies_v2.notnull())) #know null and zero elements
available.all(axis = 1)
```

```python
available.all(axis = 1).value_counts() #too much data loss
```

Remove just the 'worldwide_gross' null elements because is considered the most important column

```python
mask = available['worldwide_gross']
mask
```

```python
movies_v2 = movies_v2[mask]
```

```python
((movies_v2 != 0) & (movies_v2.notnull())).worldwide_gross.value_counts() #testing
```

Use sklearn.impute to replace the resto of the null elements in the dataset

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
```

To use 'SimpleImputer', drop the object elements

```python
movies_v2 = movies_v2.drop('movie_title', axis = 1)
```

```python
movies_v2 = movies_v2.drop('duration', axis = 1) #delete the duplicate column
movies_v2
```

```python
values = imputer.fit_transform(movies_v2) #replace the elements with the average
values
```

The numpy array is converted into a pd dataframe

```python
X = pd.DataFrame(values)
X.columns = movies_v2.columns #Use the same columns in 'movies_v2'
X.index = movies_v2.index
X.head()
```

```python
X.to_csv('datasets//intermediate_results//X.csv', index = False) #Export the clean data into a new dataset
```
